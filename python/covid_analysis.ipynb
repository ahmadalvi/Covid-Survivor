```python
df['Reported Date'] = pd.to_datetime(df['Reported Date'])
```

```python
plt.figure(figsize = (12, 7))
sns.lineplot('Reported Date', 'Confirmed Positive', data = df)

plt.title('Confirmed Positive COVID-19 Cases')
```

```python
plt.figure(figsize = (12, 7))
sns.lineplot('Reported Date', 'Deaths', data = df)
plt.title('Number of Deaths Caused By COVID-19 in Ontario')
```

```python
plt.figure(figsize = (12, 7))
sns.lineplot('Reported Date', 'Number of patients hospitalized with COVID-19', data = df)
plt.title('Number of Citizens Hospitalized with COVID-19')
```

```python
plt.figure(figsize = (12, 7))
sns.lineplot('Reported Date', 'Number of patients in ICU with COVID-19', data = df)
plt.title('Number of Patients in ICU With COVID-19')
```

```python
sns.jointplot(x = 'Confirmed Positive', y = 'Deaths', data = df)
```
